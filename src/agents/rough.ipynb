```python
from langchain.agents import create_agent
from langchain_community.agent_toolkits.polygon.toolkit import PolygonToolkit
from langchain_community.utilities.polygon import PolygonAPIWrapper
from langchain_google_genai import ChatGoogleGenerativeAI
from pydantic import BaseModel
from langchain_core.messages import BaseMessage,HumanMessage
from typing import List
from dotenv import load_dotenv
import requests
```

```python
load_dotenv()
API_KEY = "QMfJzyi0YqDhIcmxVnQ_Bec3WLqpDlgR"
polygon = PolygonAPIWrapper(polygon_api_key=API_KEY)
toolkit = PolygonToolkit.from_polygon_api_wrapper(polygon)
```

```python
class State(BaseModel):
    messages : List[BaseMessage]
    query : str
```

```python
GEMINI_API_KEY = "AIzaSyBNV8kUREIefA5OlSmNHuayTLYUKoHEuVo"
model = ChatGoogleGenerativeAI(
        model="gemini-2.5-flash",
        temperature=0,
        google_api_key=GEMINI_API_KEY
)
```

```python
agent = create_agent(
    model = model,
    tools = toolkit.get_tools(),
    state_schema=State,
    system_prompt="You are an assistant"
)
```

```python
query = "Do you have the knowledge to fetch the past records of a company?"
agent.invoke({
    "messages": [
        HumanMessage(content=query)
    ],
    "query": query
})
```
